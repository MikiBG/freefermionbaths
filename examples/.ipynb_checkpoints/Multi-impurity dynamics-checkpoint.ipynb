# Multi-impurity dynamics

This notebook shows how the code can be used to compute the time-evolution of the total impurity occupation for multiple impurities coupled equidistantly to a 1D bath, both exactly (using spectral methods) and approximately (using the formula derived from the master equation).

<!-- # <span style='background:lightgreen'> Master equation dynamics in 1D </span> -->
## Master equation dynamics in 1D

For a fully inverted state, $C(0) = I_{N\times N}$, the total impurity occupation $\langle N_e\rangle_t=\mathrm{tr}\,C(t)$ evolves as 

$$\langle N_e\rangle_t = \lVert e^{-i \Sigma(\Delta^+) t}\rVert_{\rm HS}^2\,,$$

where $\Sigma$ denotes the self-energy matrix. We recall that for a set of impurities coupled to a 1D bath,

$$\Sigma_{mn}(\Delta^+) = - i \frac{\Gamma_0}{2} e^{ik_\Delta \lvert x_m - x_n \rvert}\,.$$

```julia
using ToeplitzMatrices, LinearAlgebra

frob2(x) = sum(abs2.(x))    # Frobenius norm squared

# (positive) momentum of bath modes with frequency `Delta`
resonant_momentum(Delta) = acos(1 - Delta/2)    

"""
    times, nexcs = nexc_mastereq(N, kd, T, Nsteps)
    
Compute the master-equation evolution of the total impurity occupation for a 
set of equidistant impurities, initially occupied, coupled to a 1D empty bath.
    
#### Arguments
- `N`: number of impurities.
- `kd`: resonant momentum times the distance between neighbouring impurities.
- `T`: final time (in units of the single-impurity lifetime) until which
  the total impurity occupation is computed.
- `Nsteps`: number of steps in the evolution.

#### Outputs
- `times`: range of times (in units of the single-impurity lifetime) at which 
the total impurity occupation is evaluated.
- `nexcs`: total impurity occupation at different times.
"""
function nexc_mastereq(N, kd, T, Nsteps)
    times = range(0, T, length=(Nsteps + 1))
    dt = step(times)
    nexcs = similar(times, Float64)
    dV = exp(Array(SymmetricToeplitz([-dt/2*exp(im*kd*n) for n in 0:(N - 1)])))
    V = Matrix{ComplexF64}(I, N, N)
    nexcs[1] = frob2(V)
    for i in 2:(Nsteps + 1)
        V = dV * V
        nexcs[i] = frob2(V)
    end
    times, nexcs
end
```

```julia
Delta = 1.0
N = 5
d = 25

# evolution of impurity occupation (master equation)
times_me, record_me = nexc_mastereq(N, resonant_momentum(Delta)*d, 10.0, 500);
```

## Exact dynamics in 1D

```julia
include("../src/FiniteSizeDynamics.jl")
using .FiniteSizeDynamics
```

```julia
impurity_list = [Impurity(Delta=Delta, g=0.2, x=(1 + (n - 1)*d,)) for n in 1:N];
```

```julia
using LinearAlgebra

decay_rate(imp) = 2*imp.g^2/sqrt(4 - (imp.Delta - 2)^2)    # single-impurity decay rate
gamma = decay_rate(impurity_list[1])
dt = 0.01                  # trotterisation time step
T = times_me[end]/gamma    # final time

# initial correlation matrix (occupied impurity + Fermi sea)
L = 2^10     # number of bath sites (should be large enough to accomodate all the impurities)
Ef = -1.0    # Fermi level (empty bath)
Cc = Matrix{ComplexF64}(I, N, N)
Cb = diagm(ComplexF64[(2*(1 - cos(2*pi*k/L)) < Ef) for k in 0:(L - 1)])
Ccb = zeros(ComplexF64, N, L)

# evolution of impurity occupation (spectral method)
output(Cc, Cb, Ccb) = [real(sum(diag(Cc)))]
@time times, record = spectral_evolution!(output, Cc, Cb, Ccb, impurity_list, dt, ceil(Int, T/dt));
```

```julia
using Plots

plot(
    xlabel = "\$\\Gamma_0t\$", ylabel = "\$\\langle N_e\\rangle_t\$",
    xlim = (0, gamma*T), ylim = (0, N)
)

# plot master equation result
plot!(times_me, record_me, lw=1.5, label="master equation")

# plot spectral method result
plot!(gamma*times, record, lw=1.5, label="spectral method")
```
